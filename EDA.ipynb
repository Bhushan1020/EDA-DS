```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris = pd.read_csv(url, header=None, names=columns)

print(iris.head())
```

```python
import matplotlib.pyplot as plt

iris.hist(figsize=(10, 8))
plt.suptitle('Histograms of Iris Features')
plt.show()

sns.pairplot(iris, hue='species')
plt.suptitle('Scatter Plot Matrix of Iris Features')
plt.show()

```

```python
correlation_matrix = iris.drop(columns = 'species').corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Iris Features')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for i, column in enumerate(iris.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=column, data=iris)
    plt.title(f'Box Plot of {column}')
plt.tight_layout()
plt.show()
```

```python
def log_transform(df, column):
 df[column] = np.log(df[column] + 1)
 return df

for column in iris.columns[:-1]:
 iris = log_transform(iris, column)
```

```python
plt.figure(figsize=(12, 8))
for i, column in enumerate(iris.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=column, data=iris)
    plt.title(f'Box Plot of {column}')
plt.tight_layout()
plt.show()
```
